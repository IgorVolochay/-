```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def binary_step(num):
    if num < 0:
        return 0
    elif num >= 0:
        return 1
    
plt.plot([binary_step(i) for i in range(-10, 10)])
plt.grid()
plt.show()
```

```python
def sigmoid(num):
    return 1 / (1 + np.exp(-num))

plt.plot([sigmoid(i) for i in range(-10, 10)])
plt.grid()
plt.show()
```

```python
def tanh(num):
    return (np.exp(num) - np.exp(-num)) / (np.exp(num) + np.exp(-num))

plt.plot([tanh(i) for i in range(-10, 10)])
plt.grid()
plt.show()
```

```python
def ReLU(num):
    if num < 0:
        return 0
    elif num >= 0:
        return num
    
plt.plot([ReLU(i) for i in range(-10, 10)])
plt.grid()
plt.savefig("RelU_graf", dpi=200)
plt.show()
```

```python
def softplus(num):
    return np.log(1 + np.exp(num))

plt.plot([softplus(i) for i in range(-10, 10)])
plt.grid()
plt.savefig("softplus_graf", dpi=200)
plt.show()
```

```python
def leaky_ReLU(num):
    k = 0.2
    
    if num < 0:
        return num * k
    elif num >= 0:
        return num
    
plt.plot([leaky_ReLU(i) for i in range(-10, 10)])
plt.grid()
plt.show()
```
