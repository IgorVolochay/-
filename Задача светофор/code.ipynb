```python
weight = [0, 0, 0]
inp = [[1, 0, 0],
       [0, 1, 0],
       [0, 0, 1]]
goal_pred = [0, 0, 1]
```

```python
def neural_network(inp, weight):
    output = 0
    for i in range(len(inp)):
        output += (inp[i] * weight[i])
    return output
```

```python
alpha = 0.1
for iteration in range(50):
    for i in range(len(inp)):
        pred = neural_network(inp[i], weight)
        delta = pred - goal_pred[i]
        error = delta ** 2
        weight_delta = [0, 0, 0]
        
        for j in range(len(weight_delta)):
            weight_delta[j] = inp[i][j] * delta
            
        for j in range(len(weight)):
            weight[j] -= weight_delta[j] * alpha
            
    print(pred, weight)
```
