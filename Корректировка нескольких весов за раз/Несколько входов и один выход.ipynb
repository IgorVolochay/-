```python
weights  = [0.1, 0.49, 0.9]
inp = [8.4, 0.24, 1.1]
goal_pred = 1.5

def neural_network(inp, weights):
    output = 0
    for i in range(len(inp)):
        output += (inp[i] * weights[i])
    return output
```

```python
alpha = 0.01

for iteration in range(30):
    pred = neural_network(inp, weights)
    delta = pred - goal_pred
    error = delta ** 2
    weights_delta = [0, 0, 0]
    
    for i in range(len(weights_delta)):
        weights_delta[i] = inp[i] * delta
    
    print("Iteration: " + str(iteration + 1))
    print("Pred: " + str(pred))
    print("Error: " + str(error))
    print("Weights: " + str(weights))
    print("Weights_Deltas: " + str(weights_delta) + "\n")
    
    for i in range(len(weights)):
        weights[i] -= weights_delta[i] * alpha
```
